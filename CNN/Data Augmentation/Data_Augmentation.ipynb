```python
import matplotlib.pyplot as plt
import tensorflow
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
img = image.load_img('/content/cat.jpeg', target_size=(200,200))
```

```python
#plt.imshow(img)
img
```

```python
datagen = ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    vertical_flip=True
)
```

```python
img = image.img_to_array(img)
```

```python
print(type(img))
img.shape
```

```python
input_batch = img.reshape(1,200, 200, 3)
```

```python
import os
os.makedirs('aug', exist_ok=True)
```

```python
#.flow -> here work for single image
# .flowdirec -> something use for directory of images
i=0
for output in datagen.flow(input_batch, batch_size=1, save_to_dir='aug'):
  i=i+1
  if i==10:
    break
```
