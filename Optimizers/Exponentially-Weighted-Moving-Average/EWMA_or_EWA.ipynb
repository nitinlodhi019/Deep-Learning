```python
import numpy as pandas
import pandas as pd

import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/content/DailyDelhiClimateTest.csv')
data.head(5)
```

```python
data = data[['date', 'meantemp']]
data.head()
```

```python
plt.scatter(data['date'], data['meantemp'], color='red')
```

```python
x1 = data['meantemp'].ewm(alpha=0.9).mean()
```

```python
data.loc[:, 'ewma'] = x1
data.head()
```

```python
plt.scatter(data['date'], data['meantemp'], color='red')
plt.plot(data['date'], x1, color='black', label='alpha=0.9')
plt.show()
```
